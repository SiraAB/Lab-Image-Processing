```python
#Lab 2.1.1
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("catto.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_adjusted = img.copy()

rows = 1
cols = 2

for i in range(img.shape[1]):
    for j in range(img.shape[0]):
        a = 1
        b = 150
        f = np.float64(img[j,i])
        g = a*f + b
        g[g>255] = 255
        g[g<0] = 0
        img_adjusted[j,i] = g

plt.subplot(rows,cols,1)
plt.imshow(img)
plt.title("Original")

plt.subplot(rows,cols,2)
plt.imshow(img_adjusted)
plt.title("Adjusted")

plt.show()
```

```python
#Lab 2.1.2
import numpy as np
import cv2

img = cv2.imread("catto.jpg", cv2.IMREAD_GRAYSCALE)
frames = []

def adjust_brightness(img, b):
    a = 1
    img_adjusted = np.float64(img) * a + b
    img_adjusted[img_adjusted > 255] = 255
    img_adjusted[img_adjusted < 0] = 0
    return img_adjusted.astype(np.uint8)

start_offset = -300
end_offset = 1000
step = (end_offset - start_offset) / 100

for i in range(40):
    b = start_offset + i * step
    frame = adjust_brightness(img, b)
    frames.append(frame)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
height, width = frames[0].shape
out = cv2.VideoWriter('output2.mp4', fourcc, 2.0, (width, height), False)

for frame in frames:
    out.write(frame)

out.release()
```
