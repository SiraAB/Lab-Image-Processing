```python
# 1.2.1 Color Model Conversion Using cv2.cvtColor()

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("cat.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
img_YcrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

rows = 4
cols = 4

fig = plt.figure(figsize=(17, 12))

fig.add_subplot(rows,cols,1)
plt.imshow(img_rgb)
plt.title("RGB")

fig.add_subplot(rows,cols,2)
plt.imshow(img_rgb[:,:,0], cmap="gray")
plt.title("R")

fig.add_subplot(rows,cols,3)
plt.imshow(img_rgb[:,:,1], cmap="gray")
plt.title("G")

fig.add_subplot(rows,cols,4)
plt.imshow(img_rgb[:,:,2], cmap="gray")
plt.title("B")

fig.add_subplot(rows,cols,5)
plt.imshow(img_hsv)
plt.title("HSV")

fig.add_subplot(rows,cols,6)
plt.imshow(img_hsv[:,:,0], cmap="gray")
plt.title("H")

fig.add_subplot(rows,cols,7)
plt.imshow(img_hsv[:,:,1], cmap="gray")
plt.title("S")

fig.add_subplot(rows,cols,8)
plt.imshow(img_hsv[:,:,2], cmap="gray")
plt.title("V")

fig.add_subplot(rows,cols,9)
plt.imshow(img_hls)
plt.title("HLS")

fig.add_subplot(rows,cols,10)
plt.imshow(img_hls[:,:,0], cmap="gray")
plt.title("H")

fig.add_subplot(rows,cols,11)
plt.imshow(img_hls[:,:,1], cmap="gray")
plt.title("L")

fig.add_subplot(rows,cols,12)
plt.imshow(img_hls[:,:,2], cmap="gray")
plt.title("S")

fig.add_subplot(rows,cols,13)
plt.imshow(img_YcrCb)
plt.title("YCrCb")

fig.add_subplot(rows,cols,14)
plt.imshow(img_YcrCb[:,:,0], cmap="gray")
plt.title("Y")

fig.add_subplot(rows,cols,15)
plt.imshow(img_YcrCb[:,:,1], cmap="gray")
plt.title("Cr")

fig.add_subplot(rows,cols,16)
plt.imshow(img_YcrCb[:,:,2], cmap="gray")
plt.title("Cb")


plt.show()
```

```python
# 1.2.3 Image Bitwise AND Operation
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("cat.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)

mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:300] = 255

masked_img = cv2.bitwise_and(img, img, mask=mask)

rows = 1
cols = 3

fig = plt.figure(figsize=(17, 12))

fig.add_subplot(rows,cols,1)
plt.imshow(img)
plt.title("Original")

fig.add_subplot(rows,cols,2)
plt.imshow(mask, cmap="gray")
plt.title("Mask")

fig.add_subplot(rows,cols,3)
plt.imshow(masked_img)
plt.title("Masked Image")

plt.show()


```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import glob

img1 = cv2.imread("cat.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img_cat = cv2.resize(img1, (400, 400), interpolation=cv2.INTER_AREA)
img2 = cv2.imread("dog.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img_dog = cv2.resize(img2, (400, 400), interpolation=cv2.INTER_AREA)

frames = []

for i in range(11):
    alpha = 1 - i * 0.1
    beta = i * 0.1
    frame = cv2.addWeighted(img_cat, alpha, img_dog, beta, 0)
    frames.append(frame)
    plt.imsave(f"frame{i}.jpg", frame)

for i in range(1, 11):
    alpha = i * 0.1
    beta = 1 - i * 0.1
    frame = cv2.addWeighted(img_cat, alpha, img_dog, beta, 0)
    frames.append(frame)
    plt.imsave(f"frame{i+10}.jpg", frame)

img_array = []
for filename in glob.glob("*.jpg"):
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width, height)
    img_array.append(img)

out = cv2.VideoWriter("output.mp4", cv2.VideoWriter_fourcc(*'DIVX'), 5, size)

for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```
